```python
import csv
import pandas as pd
import random
from itertools import zip_longest

df1 = pd.DataFrame(columns=['Enrollmentno','choice1','choice2','choice3','choice4','choice5','choice6','choice7','choice8','choice9','choice10'])

df2 = pd.read_csv('studentdetails.csv')

enro= df2['Enrollmentno'].tolist()
df3 = pd.read_csv('collegedetails.csv')

colcodes= df3['CollegeCode'].tolist()

cho1=[]
cho2=[]
cho3=[]
cho4=[]
cho5=[]
cho6=[]
cho7=[]
cho8=[]
cho9=[]
cho10=[]

a1=['CSE','IT','ME']
a2=['ME','EC']
a3=['EE','IC','CH','ICT']

#chocie 1
while len(cho1) !=10000 :
    x= random.choice(colcodes)
    y = random.choice(a1)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val > 0: 
        cho1.append(x+'-'+y)

#choice 2  
while len(cho2) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a1)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho2.append(x + '-' + y)

    
#choice 3 

while len(cho3) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a2+a1)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho3.append(x + '-' + y)

    
    

#choice 4  
while len(cho4) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a2)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho4.append(x + '-' + y)
        
    
#choice 5 
while len(cho5) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a2)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho5.append(x + '-' + y)

    
 #choice 6 
while len(cho6) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a2)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho6.append(x + '-' + y)

 #choice7  
while len(cho7) !=10000:
    x=random.choice(colcodes)
    y=random.choice(a2)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val>0 :
        cho7.append(x + '-' + y)

 #choice8 
while len(cho8) !=10000 :
    x= random.choice(colcodes)
    y = random.choice(a3)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val > 0: 
        cho8.append(x+'-'+y)


 #choice9 
while len(cho9) !=10000 :
    x= random.choice(colcodes)
    y = random.choice(a3)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val > 0: 
        cho9.append(x+'-'+y)


 #choice10   
while len(cho10) !=10000 :
    x= random.choice(colcodes)
    y = random.choice(a3)
    row=df3.loc[df3['CollegeCode']==x]
    val=row[y].values[0]
    if val > 0: 
        cho10.append(x+'-'+y)
        
#overwrite
df1= pd.read_csv("collegedetails.csv")
print(df1.head())
topcoll = df1.loc[0:2,'CollegeCode']
branc=["CSE","IT","ME","ICT"]
for i in range(0,201):
    cho1[i] = random.choice(topcoll)+'-'+random.choice(branc)
    cho2[i] = random.choice(topcoll)+'-'+random.choice(branc)
    cho3[i] = random.choice(topcoll)+'-'+random.choice(branc)

fin = [enro,cho1, cho2, cho3, cho4, cho5, cho6, cho7, cho8, cho9 , cho10]
export_data = zip_longest(*fin, fillvalue = '')
with open('choiceselect.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("Enrollmentno","choice1","choice2","choice3","choice4","choice5","choice6","choice7","choice8","choice9","choice10"))
      wr.writerows(export_data)
myfile.close()

```
