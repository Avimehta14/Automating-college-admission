```python
import pandas as pd 


data = pd.read_csv("studentdetails.csv")
data = pd.DataFrame(data,columns=['Enrollmentno','sub1','sub2','sub3','sub4','sub5'])
print (data)
sum_row=data.sum(axis=1)
print(sum_row)

data["Total"]=data.sum(axis=1)

data.sort_values("Total", inplace = True,ascending=False) 

data["Rank"] = data["Total"].rank(ascending = False, method= 'first').astype(int)

data.to_csv('studentdetails.csv',index=False) 
# display 
data

 
```
