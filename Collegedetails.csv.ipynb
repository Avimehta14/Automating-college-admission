```python
import csv
import random
import pandas as pd
from itertools import zip_longest


list3=[]
s='C'
    
randomList = random.sample(range(1000, 9999),100)
rlist = randomList
while len(rlist)!=0:
    chosen = random.choice(rlist)
    list3.append('C'+str(chosen))
    rlist = list(set(rlist) - set([chosen]))
randomList    
    
    
    
    
    
list1=[0,60,120,180,240]
CSE=[]   

for i in range(0,100): 
    
    CSE.append(random.choice(list1))                                            



IT=[]   

for i in range(0,100): 
    
    IT.append(random.choice(list1)) 
    



ME=[]   

for i in range(0,100): 
   
    ME.append(random.choice(list1))                                            



EE=[]   

for i in range(0,100): 
    
    EE.append(random.choice(list1))                                            




EC=[]   

for i in range(0,100): 
    
    EC.append(random.choice(list1))                                            


IC=[]   

for i in range(0,100): 
    
    IC.append(random.choice(list1))                                            


CH=[]   

for i in range(0,100): 
    
    CH.append(random.choice(list1))                                            

ICT=[]   

for i in range(0,100): 
    ICT.append(random.choice(list1))                                            


Rank=[]

randomList = random.sample(range(1, 101),100)
rlist = randomList
while len(rlist)!=0:
    chosen = random.choice(rlist)
    Rank.append(chosen)
    rlist = list(set(rlist) - set([chosen]))
    
     
   

   
d = [list3,CSE,IT,ME,EE,EC,IC,CH,ICT,Rank]
export_data = zip_longest(*d, fillvalue = '')
with open('collegedetails.csv', 'w', encoding="ISO-8859-1", newline='') as myfile:
      wr = csv.writer(myfile)
      wr.writerow(("CollegeCode","CSE", "IT","ME","EE","EC","IC","CH","ICT","Rank"))
      wr.writerows(export_data)  
myfile.close()

df= pd.read_csv('collegedetails.csv')
nrank = df.sort_values('Rank', inplace=True)
df.to_csv('collegedetails.csv', index=False)
```
